- **Fonction rand_bi_gauss** : sert à générer des données provenant de deux variables gaussiennes bidimensionnelles. Elle prend en paramètre le nombre d'observations de chaque classe (n1 et n2), leur moyenne (mu1 et mu2) et leur écart-type (sigma1 et sigma2). - **Fonction rand_tri_gauss** : sert à générer des données provenant de trois variables gaussiennes bidimensionnelles. - **Fonction rand_clown** : sert à générer des données en forme de "clown" pour deux classes. Elle prend en paramètre le nombre d'observations de chaque classe (n1 et n2) et leur écart-type (sigma1 et sigma2). Elle retourne les labels associés à chaque observation (1 ou -1).

- **Fonction rand_checkers** : sert à générer des données en forme de "damier" pour deux classes. Elle prend en paramètre le nombre d'observations de chaque classe (n1 et n2), leur moyenne (mu1 et mu2) et une valeur optionnelle pour le bruit (sigma). Elle retourne les labels associés à chaque observation (1 ou -1).

```python
# Paramètres pour chaque fonction
params_bi_gauss = dict(n1=20, n2=20, mu1=[1, 1], mu2=[-1, -1], sigmas1=[0.9, 0.9], sigmas2=[0.9, 0.9])
params_tri_gauss = dict(n1=50, n2=50, n3=50, mu1=[1, 1], mu2=[-1, -1], mu3=[1, -1], sigma1=[0.9, 0.9], sigma2=[0.9, 0.9], sigma3=[0.9, 0.9])
params_clown = dict(n1=50, n2=50, sigma1=1, sigma2=5)
params_checkers = dict(n1=150, n2=150)

# On génère les différents de jeux de données associés à chaque fonction
data1, labels1 = rand_bi_gauss(**params_bi_gauss)
data2, labels2 = rand_tri_gauss(**params_tri_gauss)
data3, labels3 = rand_clown(**params_clown)
data4, labels4 = rand_checkers(**params_checkers)
```

## Approche intuitive

### Question 3

Essayons de transposer la méthode des k-plus proches voisins pour la classification vers la régression. Comme dans la classification, le principe de base consiste à calculer la distance entre le point à prédire et tous les points d'apprentissage. En fonction de cela, on détermine les k-plus proches voisins du point à prédire. Enfin, et c'est ici que cela va varier par rapport à la classification, nous n'allons pas catégoriser les k-plus proches voisins selon des classes, mais selon la moyenne des valeurs de y des k-plus proches voisins. Cela déterminera la prédiction de la valeur de y de notre nouveau point.

## Approche formelle

### Question 4

```python
# Classifieur k-plus proches voisins 

class KNNClassifier(BaseEstimator, ClassifierMixin):
    """ Homemade kNN classifier class """
    
    def __init__(self, n_neighbors=1):
        # Initialisation de la classe
        self.n_neighbors = n_neighbors
        self.X_train = None
        self.y_train = None
    
    def fit(self, X, y):
        # Entraînement du modèle sur les vecteurs des variables explicatives (X) et à expliquer (y)
        self.X_train = X
        self.y_train = y
        return self
    
    def predict(self, X):
        # Déclaration du vecteur de prédictions
        predictions = []

        for x in X:
            # Calcul des distances
            distances = np.linalg.norm(self.X_train - x, axis=1)

            # Indices des k-plus proches voisins
            indices = np.argsort(distances)[:self.n_neighbors]
            
            # Classes des k-plus proches voisins
            neighbors_classes = self.y_train[indices]

            # Gestion des ex-æquo en utilisant le choix aléatoire
            if len(np.unique(neighbors_classes)) == 1:
                # Choix aléatoire parmi les classes égales
                predicted_class = np.random.choice(neighbors_classes)
            else:
                # Choix de la classe majoritaire en cas de non-ex æquo
                predicted_class, _ = mode(neighbors_classes)
                
            # Ajout dans le vecteur à la fin de chaque itération
            predictions.append(predicted_class)

        return np.array(predictions)
```

```python
# On utilise le jeu de données #2
X, y = data2, labels2

# Séparation des données en ensembles d'apprentissage et de test
# Utilisation des observations d'indice pair pour l'apprentissage
X_train = X[::2] 
y_train = y[::2]

# Utilisation des observations d'indice impair pour le test
X_test = X[1::2]  
y_test = y[1::2]

# Utilisation de la classe définie précédemment KNNClassifier
knn_custom = KNNClassifier(n_neighbors=3)
knn_custom.fit(X_train, y_train)
y_pred_custom = knn_custom.predict(X_test)
# Utilisation de la classe KNeighborsClassifier de scikit-learn
knn_sklearn = KNeighborsClassifier(n_neighbors=3)
knn_sklearn.fit(X_train, y_train)
y_pred_sklearn = knn_sklearn.predict(X_test)

# Comparaison des résultats entre les deux classes
accuracy_custom = np.mean(y_pred_custom == y_test)
accuracy_sklearn = np.mean(y_pred_sklearn == y_test)

print(f"Accuracy (Homemade KNN): {accuracy_custom}")
print(f"Accuracy (Scikit-learn KNN): {accuracy_sklearn}")
```

On a un taux d'accuracy de 77,33% pour les deux méthodes de KNN (celle que nous avons codé, et celle de Scikit-learn). Cela signifie que les prédictions sont cohérentes pour 77,33% des cas.

```python
# Chargement des données du jeu de données #2 (observations d'indice pair uniquement)
X_train = data2[::2]
y_train = labels2[::2]

# Création d'un modèle kNN avec différentes valeurs de k
k_values = range(1, 13)
plt.figure(figsize=(15, 10))

for i, k in enumerate(k_values):
    plt.subplot(3, 4, i + 1)
    
    # Création et entraînement le modèle kNN sur nos données d'entraînement (observations d'indice pair)
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    # Affichage de la frontière de décision
    frontiere_new(lambda x: knn.predict([x]), X_train, y_train)
    plt.title(f'k = {k}')

# Affichage des graphiques    
plt.tight_layout()
plt.show()
```

- On remarque que, dans le **cas k=1**, l'algorithme attribue le label du seul voisin le plus proche au point de test. Cela signifie que chaque point de test est attribué à la classe de son voisin le plus proche. Cela résulte en une frontière de décision est très complexe, car elle est très sensible à la moindre variation locale des données. Cette frontière est très irrégulière, et peut souffrir d'un surajustement aux données d'apprentissage. 

- Dans le cas **k=n** (ici, k=12), chaque point de test est influencé par l'intégralité de l'ensemble d'apprentissage. La frontière de décision devient donc très lisse et régulière, car elle est issue de la moyenne de tous les labels des points d'apprentissage. Cela la rend donc moins sensible aux variations locales, ce qui explique la simplicité de la frontière de décision.

- La frontière est donc **complexe dans le cas k=1** et **simple dans le cas k=n**.

### Question 11

**Avantages de la méthode des k-plus proches voisins**

- *Interprétabilité* : Les prédictions sont assez intuitives à comprendre, car elles sont basées sur les étiquettes des voisins les plus proches.
- *Facilité d'implémentation* : Le k-NN est conceptuellement simple et facile à mettre en œuvre, ce qui le rend attrayant pour de nombreux problèmes de classification.
- *Pas de phase d'apprentissage* : Contrairement à d'autres algorithmes, k-NN n'a pas de phase d'apprentissage explicite. Il suffit de stocker les données d'entraînement, rendant le processus d'entraînement extrêmement rapide.
- *Flexibilité* : Le nombre de voisins (k) peut être ajusté pour améliorer les performances du modèle. Un réglage approprié de k peut aider à équilibrer le biais et la variance.

**Inconvénients de la méthode des k-plus proches voisins**
- *Temps de calcul élevé* : La nécessité de calculer la distance de chaque point de données de l'ensemble d'entraînement à chaque point de l'ensemble de test rend le k-NN lent pour les grands ensembles de données.
- *Sensibilité aux données de faible qualité* : Le k-NN est sensible au bruit, aux valeurs aberrantes et aux données non pertinentes, car toutes les caractéristiques contribuent de manière égale à la distance.
- *Mauvais passage à l'échelle* : Avec l'augmentation du nombre de caractéristiques (malédiction de la dimensionnalité), la performance du k-NN peut se dégrader, car il devient difficile de trouver des voisins significatifs dans des espaces de grande dimension.

```python
# Classifieur LDA (Linear Discriminant Analysis, ou Analyse Discriminante Linéaire)

class LDAClassifier:
    """ Homemade LDA classifier class """
    
    def __init__(self):
        # Initialisation de la classe
        self.mu_plus = None
        self.mu_minus = None
        self.sigma = None
        self.pi_plus = None

    def fit(self, X, y):
        # Entraînement du modèle sur les vecteurs des variables explicatives (X) et à expliquer (y) + Estimation des paramètres
        n = X.shape[0]
        m = np.sum(y == 1)
        
        self.pi_plus = m / n
        self.mu_plus = np.mean(X[y == 1], axis=0)
        self.mu_minus = np.mean(X[y == -1], axis=0)

        sigma_plus = np.cov(X[y == 1].T)
        sigma_minus = np.cov(X[y == -1].T)
        self.sigma = ((m - 1) * sigma_plus + (n - m - 1) * sigma_minus) / (n - 2)

    def predict(self, X):
        # Calcul de la règle de décision
        inv_sigma = np.linalg.inv(self.sigma)
        w = np.dot(inv_sigma, self.mu_plus - self.mu_minus)
        b = -0.5 * np.dot(self.mu_plus.T, np.dot(inv_sigma, self.mu_plus)) + \
            0.5 * np.dot(self.mu_minus.T, np.dot(inv_sigma, self.mu_minus)) + \
            np.log(1 - self.pi_plus) - np.log(self.pi_plus)

        return np.where(np.dot(X, w) > b, 1, -1)
```

```python
# Génération des données déséquilibrées grâce à la fonction rand_bi_gauss
data_bi_gauss, labels_bi_gauss = rand_bi_gauss(n1=100, n2=20, mu1=[1, 1], mu2=[-1, -1], sigmas1=[0.9, 0.9], sigmas2=[0.9, 0.9])

# Séparation des données en ensembles d'entraînement (X_train, y_train) et de test (X_test, y_test)
X_train, X_test, y_train, y_test = train_test_split(data_bi_gauss, labels_bi_gauss, test_size=0.3, random_state=42)

# Création d'une instance de la classe LogisticRegression
log_reg = LogisticRegression()

# Entraînement de la régression logistique sur les données d'entraînement (X_train, y_train)
log_reg.fit(X_train, y_train)

# Entraînement de la LDA sur les données d'entraînement (X_train, y_train)
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)

# Prédiction et précision de la régression logistique
predictions_lr = log_reg.predict(X_test)
accuracy_lr = accuracy_score(y_test, predictions_lr)

# Prédiction et précision de la LDA
predictions_lda = lda.predict(X_test)
accuracy_lda = accuracy_score(y_test, predictions_lda)

# Affichage de la précision pour chacune des deux méthodes
print(f"Précision Régression Logistique: {accuracy_lr}")
print(f"Précision LDA: {accuracy_lda}")
```

La régression logistique gère mieux les classes déséquilibrées que la LDA, en particulier lorsqu'elle est configurée avec des paramètres appropriés (en ajustant le paramètre class_weight).

### Question 22

- **coef_** représente les coefficients des variables explicatives (features). Ces coefficients sont utilisés pour calculer la combinaison linéaire des caractéristiques qui entre dans la fonction logistique.
- **intercept_** représente le terme constant (également appelé biais ou intercept) dans la fonction de décision. C'est le point où la fonction de décision (la combinaison linéaire des caractéristiques pondérée par les coefficients) coupe l'axe des ordonnées lorsque toutes les variables explicatives sont égales à zéro.
